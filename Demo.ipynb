```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


# Load the Iris dataset (replace with your data if needed)
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]  # Add species column

# === TASK 1: Data Exploration ===
print("=== Data Exploration ===")
print("First 5 rows:")
display(data.head())  # Shows first 5 rows with formatting

print("\nDataset info:")
print(data.info())  # Checks data types and missing values

# === TASK 2: Data Analysis ===
print("\n=== Data Analysis ===")
print("Descriptive statistics:")
print(data.describe())  # Mean, median, etc.

print("\nGrouped analysis (mean by species):")
print(data.groupby('species').mean())  # Compare averages across categories

# === TASK 3: Visualizations ===
plt.figure(figsize=(15, 10))  # Set figure size for all plots

# Plot 1: Line Chart (Trends)
plt.subplot(2, 2, 1)
plt.plot(data['sepal length (cm)'], color='blue', label='Sepal Length')
plt.title("Sepal Length Trend Across Samples")
plt.xlabel("Observation #")
plt.ylabel("Length (cm)")
plt.grid(True)

# Plot 2: Bar Chart (Categories)
plt.subplot(2, 2, 2)
data.groupby('species')['petal length (cm)'].mean().plot(
    kind='bar', 
    color=['red', 'green', 'blue']
)
plt.title("Average Petal Length by Species")
plt.ylabel("Length (cm)")

# Plot 3: Histogram (Distribution)
plt.subplot(2, 2, 3)
plt.hist(data['sepal width (cm)'], bins=15, color='orange', edgecolor='black')
plt.title("Distribution of Sepal Width")
plt.xlabel("Width (cm)")
plt.ylabel("Frequency")

# Plot 4: Scatter Plot (Relationships)
plt.subplot(2, 2, 4)
for species in ['setosa', 'versicolor', 'virginica']:
    subset = data[data['species'] == species]
    plt.scatter(
        subset['sepal length (cm)'], 
        subset['petal length (cm)'], 
        label=species,
        alpha=0.7
    )
plt.title("Sepal vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()

plt.tight_layout()  # Prevent overlapping
plt.show()  # Display all plots# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset (replace with your data if needed)
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]  # Add species column

# === TASK 1: Data Exploration ===
print("=== Data Exploration ===")
print("First 5 rows:")
display(data.head())  # Shows first 5 rows with formatting

print("\nDataset info:")
print(data.info())  # Checks data types and missing values

# === TASK 2: Data Analysis ===
print("\n=== Data Analysis ===")
print("Descriptive statistics:")
print(data.describe())  # Mean, median, etc.

print("\nGrouped analysis (mean by species):")
print(data.groupby('species').mean())  # Compare averages across categories

# === TASK 3: Visualizations ===
plt.figure(figsize=(15, 10))  # Set figure size for all plots

# Plot 1: Line Chart (Trends)
plt.subplot(2, 2, 1)
plt.plot(data['sepal length (cm)'], color='blue', label='Sepal Length')
plt.title("Sepal Length Trend Across Samples")
plt.xlabel("Observation #")
plt.ylabel("Length (cm)")
plt.grid(True)

# Plot 2: Bar Chart (Categories)
plt.subplot(2, 2, 2)
data.groupby('species')['petal length (cm)'].mean().plot(
    kind='bar', 
    color=['red', 'green', 'blue']
)
plt.title("Average Petal Length by Species")
plt.ylabel("Length (cm)")

# Plot 3: Histogram (Distribution)
plt.subplot(2, 2, 3)
plt.hist(data['sepal width (cm)'], bins=15, color='orange', edgecolor='black')
plt.title("Distribution of Sepal Width")
plt.xlabel("Width (cm)")
plt.ylabel("Frequency")

# Plot 4: Scatter Plot (Relationships)
plt.subplot(2, 2, 4)
for species in ['setosa', 'versicolor', 'virginica']:
    subset = data[data['species'] == species]
    plt.scatter(
        subset['sepal length (cm)'], 
        subset['petal length (cm)'], 
        label=species,
        alpha=0.7
    )
plt.title("Sepal vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()

plt.tight_layout()  # Prevent overlapping
plt.show()  # Display all plots# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset (replace with your data if needed)
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]  # Add species column

# === TASK 1: Data Exploration ===
print("=== Data Exploration ===")
print("First 5 rows:")
display(data.head())  # Shows first 5 rows with formatting

print("\nDataset info:")
print(data.info())  # Checks data types and missing values

# === TASK 2: Data Analysis ===
print("\n=== Data Analysis ===")
print("Descriptive statistics:")
print(data.describe())  # Mean, median, etc.

print("\nGrouped analysis (mean by species):")
print(data.groupby('species').mean())  # Compare averages across categories

# === TASK 3: Visualizations ===
plt.figure(figsize=(15, 10))  # Set figure size for all plots

# Plot 1: Line Chart (Trends)
plt.subplot(2, 2, 1)
plt.plot(data['sepal length (cm)'], color='blue', label='Sepal Length')
plt.title("Sepal Length Trend Across Samples")
plt.xlabel("Observation #")
plt.ylabel("Length (cm)")
plt.grid(True)

# Plot 2: Bar Chart (Categories)
plt.subplot(2, 2, 2)
data.groupby('species')['petal length (cm)'].mean().plot(
    kind='bar', 
    color=['red', 'green', 'blue']
)
plt.title("Average Petal Length by Species")
plt.ylabel("Length (cm)")

# Plot 3: Histogram (Distribution)
plt.subplot(2, 2, 3)
plt.hist(data['sepal width (cm)'], bins=15, color='orange', edgecolor='black')
plt.title("Distribution of Sepal Width")
plt.xlabel("Width (cm)")
plt.ylabel("Frequency")

# Plot 4: Scatter Plot (Relationships)
plt.subplot(2, 2, 4)
for species in ['setosa', 'versicolor', 'virginica']:
    subset = data[data['species'] == species]
    plt.scatter(
        subset['sepal length (cm)'], 
        subset['petal length (cm)'], 
        label=species,
        alpha=0.7
    )
plt.title("Sepal vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()

plt.tight_layout()  # Prevent overlapping
plt.show()  # Display all plots# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset (replace with your data if needed)
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]  # Add species column

# === TASK 1: Data Exploration ===
print("=== Data Exploration ===")
print("First 5 rows:")
display(data.head())  # Shows first 5 rows with formatting

print("\nDataset info:")
print(data.info())  # Checks data types and missing values

# === TASK 2: Data Analysis ===
print("\n=== Data Analysis ===")
print("Descriptive statistics:")
print(data.describe())  # Mean, median, etc.

print("\nGrouped analysis (mean by species):")
print(data.groupby('species').mean())  # Compare averages across categories

# === TASK 3: Visualizations ===
plt.figure(figsize=(15, 10))  # Set figure size for all plots

# Plot 1: Line Chart (Trends)
plt.subplot(2, 2, 1)
plt.plot(data['sepal length (cm)'], color='blue', label='Sepal Length')
plt.title("Sepal Length Trend Across Samples")
plt.xlabel("Observation #")
plt.ylabel("Length (cm)")
plt.grid(True)

# Plot 2: Bar Chart (Categories)
plt.subplot(2, 2, 2)
data.groupby('species')['petal length (cm)'].mean().plot(
    kind='bar', 
    color=['red', 'green', 'blue']
)
plt.title("Average Petal Length by Species")
plt.ylabel("Length (cm)")

# Plot 3: Histogram (Distribution)
plt.subplot(2, 2, 3)
plt.hist(data['sepal width (cm)'], bins=15, color='orange', edgecolor='black')
plt.title("Distribution of Sepal Width")
plt.xlabel("Width (cm)")
plt.ylabel("Frequency")

# Plot 4: Scatter Plot (Relationships)
plt.subplot(2, 2, 4)
for species in ['setosa', 'versicolor', 'virginica']:
    subset = data[data['species'] == species]
    plt.scatter(
        subset['sepal length (cm)'], 
        subset['petal length (cm)'], 
        label=species,
        alpha=0.7
    )
plt.title("Sepal vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()

plt.tight_layout()  # Prevent overlapping
plt.show()  # Display all plots



# === TASK 2: Data Analysis ===
print("\n=== Data Analysis ===")
print("Descriptive statistics:")
print(data.describe())  # Mean, median, etc.

print("\nGrouped analysis (mean by species):")
print(data.groupby('species').mean())  # Compare averages across categories



# === TASK 3: Visualizations ===
plt.figure(figsize=(15, 10))  # Set figure size

# Plot 1: Line Chart
plt.subplot(2, 2, 1)
plt.plot(data['sepal length (cm)'], color='blue', label='Sepal Length')
plt.title("Sepal Length Trend")
plt.xlabel("Observation #")
plt.ylabel("Length (cm)")

# Plot 2: Bar Chart
plt.subplot(2, 2, 2)
data.groupby('species')['petal length (cm)'].mean().plot(kind='bar', color=['red', 'green', 'blue'])
plt.title("Avg Petal Length by Species")
plt.ylabel("Length (cm)")

# Plot 3: Histogram
plt.subplot(2, 2, 3)
plt.hist(data['sepal width (cm)'], bins=15, color='orange', edgecolor='black')
plt.title("Sepal Width Distribution")
plt.xlabel("Width (cm)")
plt.ylabel("Frequency")

# Plot 4: Scatter Plot
plt.subplot(2, 2, 4)
for species in ['setosa', 'versicolor', 'virginica']:
    subset = data[data['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['petal length (cm)'], label=species, alpha=0.7)
plt.title("Sepal vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()

plt.tight_layout()  # Prevent overlap
plt.show()
```
